```python
# ==== OPTION A: Refit emissions by EM (with gentle shrinkage & consistent state order) ====

import os
import numpy as np
import pandas as pd
import joblib
from tqdm import tqdm
from hmmlearn.hmm import GaussianHMM

# ----------------------------- CONFIG ---------------------------------
SIG_COL   = 'force_corrected (pN)'
ORDER_DESC = True      # True => State 1 = highest force (descending). False => lowest first.
SHRINK_TO_BASE = 0.20  # 0.0 = no shrink; 0.2 pulls 20% toward base after EM
MAX_ITER  = 25         # EM iterations
TOL       = 1e-3
VERBOSE   = True
OUTROOT   = "new_models"

# ------------------------ BASIC HELPERS --------------------------------
def to_col(x):
    x = np.asarray(x).reshape(-1,1)
    return x[~np.isnan(x[:,0])].reshape(-1,1)

def pack_traces(traces, col=SIG_COL):
    Xs, lengths = [], []
    for df in tqdm(traces, desc="Packing traces", leave=False):
        x = to_col(df[col].values)
        if x.size == 0: 
            continue
        Xs.append(x)
        lengths.append(len(x))
    if not Xs:
        raise ValueError("No data after NaN filtering.")
    return np.vstack(Xs), lengths

def sort_states_by_mean_safe(hmm, descending=ORDER_DESC):
    """
    Reorder states by emission mean (descending if True), and permute all params.
    """
    order = np.argsort(hmm.means_.ravel())
    if descending:
        order = order[::-1]
    hmm.means_      = hmm.means_[order]
    # Safely assign covars for hmmlearn versions
    if hasattr(hmm, "_covars_"):
        hmm._covars_ = hmm._covars_[order]
    else:
        hmm.covars_  = hmm.covars_[order]
    hmm.startprob_  = hmm.startprob_[order]
    hmm.transmat_   = hmm.transmat_[order][:, order]
    return order

def emissions_summary_df(model, label):
    means = model.means_.ravel()
    if hasattr(model, "_covars_"):
        covs = np.array(model._covars_).ravel()
    else:
        covs = np.array(model.covars_).ravel()
    order = np.arange(1, len(means)+1)  # 1..K after sorting
    return pd.DataFrame({
        "model": label,
        "state": order,
        "mean_force": means,
        "var": covs
    })

# ------------------- CORE: Adapt emissions via EM ----------------------
def adapt_emissions_with_em(
    traces,
    base_model: GaussianHMM,
    col=SIG_COL,
    update_means=True,
    update_covars=True,
    update_kinetics=True,
    max_iter=MAX_ITER,
    tol=TOL,
    shrink_to_base=SHRINK_TO_BASE,
    random_state=0,
    verbose=VERBOSE,
):
    """
    Refit a copy of base_model on new data, allowing emissions to change.
    - update_means=True lets μ_k move
    - update_covars optionally lets σ_k^2 move (start False; enable later if needed)
    - update_kinetics=True updates startprob/transmat
    After EM, μ_new can be shrunk towards μ_base for comparability.
    """
    # 1) Pack data
    X, lengths = pack_traces(traces, col=col)

    # 2) Decide which params EM updates
    params = ""
    if update_kinetics: params += "st"
    if update_means:    params += "m"
    if update_covars:   params += "c"

    # 3) Initialize from base (no reinit)
    m = GaussianHMM(
        n_components=base_model.n_components,
        covariance_type=base_model.covariance_type,
        n_iter=max_iter, tol=tol, random_state=random_state,
        verbose=verbose, params=params, init_params=''
    )
    m.means_     = base_model.means_.copy()
    if hasattr(base_model, "_covars_"):
        m._covars_ = base_model._covars_.copy()
    else:
        m.covars_  = base_model.covars_.copy()
    m.startprob_ = base_model.startprob_.copy()
    m.transmat_  = base_model.transmat_.copy()

    # 4) EM fit
    m.fit(X, lengths)

    # 5) Sort by mean (enforce consistent state identity)
    sort_states_by_mean_safe(m, descending=ORDER_DESC)

    import copy  # put at the top of the file

    if shrink_to_base and update_means:
        # Align base to the same order first (so “State 1” matches “State 1”)
        base_copy = copy.deepcopy(base_model)
        sort_states_by_mean_safe(base_copy, descending=ORDER_DESC)
        m.means_ = (1.0 - shrink_to_base) * m.means_ + shrink_to_base * base_copy.means_
        # keep order consistent after shrink
        sort_states_by_mean_safe(m, descending=ORDER_DESC)


    return m

# ------------------------ RUN FOR ALL CONDITIONS -----------------------
# Your trace lists must already be defined in the environment:
#   ctcf_traces, pds5a_traces, coh_traces, pc_traces
# (like you showed earlier)

# 0) Load the CTCF-only base emissions model (trusted baseline)
ctcf_base = joblib.load("new_models/pdgfra_nterm_ctcfwt_models/model_4_states.pkl")
sort_states_by_mean_safe(ctcf_base, descending=ORDER_DESC)  # ensure order consistency

# 1) Fit adapted models (emissions + kinetics) for each condition
os.makedirs(OUTROOT, exist_ok=True)

def run_condition(traces, base_model, out_path, label,
                  update_means=True, update_covars=False, update_kinetics=True):
    print(f"\n[{label}] Adapting emissions by EM...")
    model_adapt = adapt_emissions_with_em(
        traces, base_model,
        update_means=update_means, update_covars=update_covars, update_kinetics=update_kinetics
    )
    joblib.dump(model_adapt, out_path)
    print(f"Saved → {out_path}")
    # write quick emissions summary
    emiss_df = emissions_summary_df(model_adapt, label)
    csv_path = out_path.replace(".pkl", "_emissions.csv")
    emiss_df.to_csv(csv_path, index=False)
    print(f"Emissions summary → {csv_path}")
    print(emiss_df.to_string(index=False))
    return model_adapt

ctcf_model_adapt = run_condition(
    ctcf_traces, ctcf_base,
    out_path=os.path.join(OUTROOT, "ctcf_emissions_refit.pkl"),
    label="CTCF (EM emissions)"
)

pds5a_model_adapt = run_condition(
    pds5a_traces, ctcf_base,
    out_path=os.path.join(OUTROOT, "pds5a_emissions_refit.pkl"),
    label="PDS5A (EM emissions)"
)

cohesin_model_adapt = run_condition(
    coh_traces, ctcf_base,
    out_path=os.path.join(OUTROOT, "cohesin_emissions_refit.pkl"),
    label="Cohesin (EM emissions)"
)

pds5a_coh_model_adapt = run_condition(
    pc_traces, ctcf_base,
    out_path=os.path.join(OUTROOT, "pds5a_cohesin_emissions_refit.pkl"),
    label="PDS5A+Coh (EM emissions)"
)

# 2) (Optional) Light kinetics polish only, starting from the adapted models
#    — often not needed, because EM above already updated kinetics.
from hmmlearn.hmm import GaussianHMM

def refine_transitions_only(traces, init_model, col=SIG_COL, max_iter=5, tol=1e-3, random_state=0):
    X, lengths = pack_traces(traces, col=col)
    m = GaussianHMM(
        n_components=init_model.n_components, covariance_type=init_model.covariance_type,
        n_iter=max_iter, tol=tol, random_state=random_state,
        verbose=True, params='st', init_params=''
    )
    # emissions frozen from adapted model
    m.means_     = init_model.means_.copy()
    if hasattr(init_model, "_covars_"):
        m._covars_ = init_model._covars_.copy()
    else:
        m.covars_  = init_model.covars_.copy()
    # start/trans as adaptation starting point
    m.startprob_ = init_model.startprob_.copy()
    m.transmat_  = init_model.transmat_.copy()
    m.fit(X, lengths)
    sort_states_by_mean_safe(m, descending=ORDER_DESC)
    return m

# Example (uncomment if you want kinetics re‑polish):
# pds5a_model_adapt_refined = refine_transitions_only(pds5a_traces, pds5a_model_adapt, max_iter=5)
# joblib.dump(pds5a_model_adapt_refined, os.path.join(OUTROOT, "pds5a_emissions_refit_transitions_refined.pkl"))
```

```python
# ==== END-TO-END with stable regime labeling (ΔLL threshold + min run length) ====

import os
import numpy as np
import pandas as pd
import joblib
from tqdm import tqdm
import matplotlib.pyplot as plt

# ----------------------------- CONFIG ---------------------------------
# Windowing in *samples*. Use longer windows for fewer switches.
WIN  = 500         # try 500–1000 for stability
STEP = 100         # ~10–20% of WIN is a good start

# Stability knobs
DELTA_LL_SWITCH = 8.0   # require ≥ this ΔLL(best - current) to switch (tune 5–12)
MIN_WINDOWS_RUN = 3     # merge runs shorter than this many windows

OUTROOT = "new_models"   # where outputs go
TIME_COL = "time (seconds)"
SIG_COL  = "force_corrected (pN)"

# ------------------------ BASIC HELPERS --------------------------------
def to_col(x):
    x = np.asarray(x).reshape(-1,1)
    return x[~np.isnan(x[:,0])].reshape(-1,1)

def loglik(model, df, col=SIG_COL):
    X = to_col(df[col].values)
    return float(model.score(X))

# -------------------- SCORING (ALLOWED MODELS) -------------------------
def score_traces_against_allowed(traces, model_dict, allowed, name_prefix, outdir=OUTROOT):
    """
    Scores each trace only against models in `allowed` (must include 'CTCF').
    Saves CSV with per-trace log-likelihoods and best model.
    """
    os.makedirs(outdir, exist_ok=True)
    keep = {k: model_dict[k] for k in allowed}
    rows = []
    for i, df in enumerate(tqdm(traces, desc=f"Scoring {name_prefix}")):
        scores = {mname: loglik(m, df) for mname, m in keep.items()}
        best = max(scores, key=scores.get)
        dll_vs_ctcf = scores[best] - scores["CTCF"]  # assumes 'CTCF' in allowed
        rows.append({"trace_idx": i, **scores, "best_model": best, "dLL_vs_CTCF": dll_vs_ctcf})
    out = pd.DataFrame(rows)
    path = os.path.join(outdir, f"{name_prefix}_trace_scores.csv")
    out.to_csv(path, index=False)
    print(f"Saved → {path}")
    return out

# --------------- SLIDING WINDOWS (MULTI-MODEL) ------------------------
def sliding_multi_ll(df, models_dict, allowed, col=SIG_COL, win=WIN, step=STEP):
    """
    Returns: windows = [(start_idx, end_idx, {model_name: LL, ...}), ...]
    """
    X = to_col(df[col].values)
    n = len(X)
    wins = []
    keep = {k: models_dict[k] for k in allowed}
    for s in range(0, max(n - win + 1, 1), step):
        e = min(s + win, n)
        seg = X[s:e]
        scores = {name: m.score(seg) for name, m in keep.items()}
        wins.append((s, e, scores))
    return wins

# ---------------- STABILITY: LABELING WITH HYSTERESIS ------------------
def _collapse_single_spikes(raw_labels):
    """
    raw_labels: [(s,e,label), ...] window-level labels (argmax-like)
    Remove A,B,A single-window spikes -> A,A,A.
    """
    if not raw_labels:
        return []
    labels = []
    for i,(s,e,lab) in enumerate(raw_labels):
        if i==0 or i==len(raw_labels)-1:
            labels.append((s,e,lab))
        else:
            prev = raw_labels[i-1][2]; nxt = raw_labels[i+1][2]
            labels.append((s,e, prev if (lab != prev and prev == nxt) else lab))
    return labels

def label_windows_multiclass_stable(windows, delta_ll_switch=DELTA_LL_SWITCH):
    """
    Hysteresis rule: start with best model; only switch if new-best beats current by ≥ delta_ll_switch.
    """
    if not windows:
        return []
    s0, e0, sc0 = windows[0]
    current = max(sc0, key=sc0.get)
    out = [(s0, e0, current)]
    for (s,e,sc) in windows[1:]:
        best = max(sc, key=sc.get)
        if (sc[best] - sc[current]) >= float(delta_ll_switch):
            current = best
        out.append((s,e,current))
    return _collapse_single_spikes(out)

def enforce_min_regime_length(labels, min_windows=MIN_WINDOWS_RUN):
    """
    Merge runs shorter than `min_windows` into adjacent regime.
    labels: [(s,e,label), ...] at the window level.
    """
    if not labels:
        return []

    # identify runs
    runs = []
    i = 0
    while i < len(labels):
        j = i + 1
        while j < len(labels) and labels[j][2] == labels[i][2]:
            j += 1
        runs.append((i, j, labels[i][2]))  # [i, j)
        i = j

    # assign labels, merging shorts
    lab_by_idx = [lab for (_,_,lab) in labels]
    for start, end, lab in runs:
        run_len = end - start
        if run_len < min_windows:
            if start > 0:
                new_lab = lab_by_idx[start-1]
            elif end < len(lab_by_idx):
                new_lab = lab_by_idx[end]
            else:
                new_lab = lab
            for k in range(start, end):
                lab_by_idx[k] = new_lab

    # rebuild
    out = []
    s_acc, e_acc, cur = labels[0][0], labels[0][1], lab_by_idx[0]
    for idx, (s,e,_) in enumerate(labels[1:], start=1):
        if lab_by_idx[idx] == cur:
            e_acc = e
        else:
            out.append((s_acc, e_acc, cur))
            s_acc, e_acc, cur = s, e, lab_by_idx[idx]
    out.append((s_acc, e_acc, cur))
    return out

# --------------- ANNOTATE ONE TRACE (stable) --------------------------
def annotate_trace_regime_multiclass(df, models_dict, allowed, col=SIG_COL,
                                     win=WIN, step=STEP,
                                     delta_ll_switch=DELTA_LL_SWITCH,
                                     min_windows=MIN_WINDOWS_RUN):
    """
    Annotates a single trace with per-sample 'regime' among `allowed` models,
    using ΔLL hysteresis + minimum run length enforcement.
    """
    wins = sliding_multi_ll(df, models_dict, allowed, col=col, win=win, step=step)
    labs = label_windows_multiclass_stable(wins, delta_ll_switch=delta_ll_switch)
    labs = enforce_min_regime_length(labs, min_windows=min_windows)

    # Expand to per-sample labels (align to non‑NaN samples only)
    mask = df[col].notna()
    n = mask.sum()
    sample_labels = np.array(['CTCF'] * n, dtype=object)
    for (s,e,lab) in labs:
        sample_labels[s:min(e, n)] = lab

    out = df.copy()
    out.loc[mask, 'regime'] = sample_labels
    return out, wins, labs

# ---------------------- PLOTTING ΔLL QUICKLOOK -------------------------
# ---------------------- PLOTTING ΔLL QUICKLOOK (robust) -------------------------
def plot_dll_quicklook(
    df, win_scores, win_labels, allowed,
    save_path=None, col=SIG_COL,
    colors=None
):
    """
    Plot ΔLL(best - baseline) over time with shaded non-baseline windows.
    Baseline defaults to 'CTCF' if present in `allowed`, else the first model in `allowed`.
    """
    if not win_scores:
        return

    # Choose baseline model
    baseline = "CTCF" if ("CTCF" in allowed) else (allowed[0] if len(allowed) else None)
    if baseline is None:
        return

    # Build time array for the *non-NaN* samples used during scoring
    t_full = df[TIME_COL].to_numpy()
    sig_full = df[col].to_numpy()
    data_mask = ~np.isnan(sig_full) & ~np.isnan(t_full)
    t = t_full[data_mask]
    if t.size == 0:
        return

    # Safe index->time mapping (win indices are in the filtered space)
    def idx_to_time(i):
        j = int(round(i))
        if j < 0: j = 0
        if j >= len(t): j = len(t) - 1
        return t[j]

    # Default colors for shading (can be overridden)
    if colors is None:
        colors = {"PDS5A":"#7db8ff", "Cohesin":"#ffa657", "PDS5A+Coh":"#c07dff", "CTCF":"#cccccc"}

    # Build ΔLL series at window centers
    centers_t, dll = [], []
    for (s, e, sc) in win_scores:
        if not sc:
            continue
        best = max(sc, key=sc.get)
        base_key = baseline if baseline in sc else best  # fallback safety
        centers_t.append(idx_to_time(0.5*(s + e)))
        dll.append(sc[best] - sc[base_key])

    if not centers_t:
        return

    # Plot
    plt.figure(figsize=(8, 3))
    plt.plot(centers_t, dll, lw=1)
    plt.axhline(0, ls='--', lw=1)

    # Shade non-baseline regimes
    used_labels = set()
    for (s, e, lab) in win_labels:
        if lab != baseline:
            t0 = idx_to_time(s)
            t1 = idx_to_time(e - 1)  # e is exclusive; shade up to last in-window sample
            c = colors.get(lab, None)
            patch = plt.axvspan(t0, t1, alpha=0.18, color=c, label=(lab if lab not in used_labels else None))
            used_labels.add(lab)

    # Legend (only if any non-baseline shading occurred)
    if used_labels:
        handles, labels = plt.gca().get_legend_handles_labels()
        # Deduplicate legend entries
        uniq = dict(zip(labels, handles))
        # Put baseline info in title, keep legend for factor regimes
        plt.legend(uniq.values(), uniq.keys(), title="Regime", loc="upper right", fontsize=8)

    plt.xlabel('time (s)')
    plt.ylabel(f'ΔLL (best − {baseline})')
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=200)
        plt.close()
    else:
        plt.show()


# ---------------- RUNNER: ANNOTATE + PLOT PER CONDITION ----------------
def run_condition_annotation_multiclass(condition_name, traces, models_dict, allowed, outdir,
                                        col=SIG_COL, win=WIN, step=STEP,
                                        delta_ll_switch=DELTA_LL_SWITCH, min_windows=MIN_WINDOWS_RUN):
    os.makedirs(outdir, exist_ok=True)
    os.makedirs(os.path.join(outdir, "annotated_csv"), exist_ok=True)
    os.makedirs(os.path.join(outdir, "plots"), exist_ok=True)

    summary = []
    for idx, df in enumerate(tqdm(traces, desc=f"{condition_name}: annotate+plot")):
        annotated_df, win_scores, win_labels = annotate_trace_regime_multiclass(
            df, models_dict, allowed, col=col, win=win, step=step,
            delta_ll_switch=delta_ll_switch, min_windows=min_windows
        )
        csv_path = os.path.join(outdir, "annotated_csv", f"{condition_name}_trace_{idx}_annotated.csv")
        annotated_df.to_csv(csv_path, index=False)

        fig_path = os.path.join(outdir, "plots", f"{condition_name}_trace_{idx}_dll.png")
        plot_dll_quicklook(annotated_df, win_scores, win_labels, allowed, save_path=fig_path, col=col)

        # summary stats
        mask = annotated_df[col].notna()
        regimes = annotated_df.loc[mask, 'regime'].to_numpy()
        switches = int(np.sum(regimes[1:] != regimes[:-1])) if regimes.size > 1 else 0
        frac = {name: (float(np.mean(regimes==name)) if regimes.size else np.nan) for name in allowed}
        summary.append({"condition": condition_name, "trace_idx": idx, "n_samples": int(mask.sum()),
                        "switches": switches, **{f"frac_{k}": v for k,v in frac.items()},
                        "annotated_csv": csv_path, "plot_png": fig_path})

    summary_df = pd.DataFrame(summary)
    summary_df.to_csv(os.path.join(outdir, f"{condition_name}_summary.csv"), index=False)
    print(f"[{condition_name}] Summary saved.")
    return summary_df

# -------------------- LOAD MODELS (use emissions-refit) ----------------
# If you prefer the older transitions_refined models, just change the filenames back.
ctcf_model      = joblib.load("new_models/ctcf_emissions_refit.pkl")
pds5a_model     = joblib.load("new_models/pds5a_emissions_refit.pkl")
cohesin_model   = joblib.load("new_models/cohesin_emissions_refit.pkl")
pds5a_coh_model = joblib.load("new_models/pds5a_cohesin_emissions_refit.pkl")

model_dict = {
    "CTCF": ctcf_model,
    "PDS5A": pds5a_model,
    "Cohesin": cohesin_model,
    "PDS5A+Coh": pds5a_coh_model,
}

# ---------------------- CHECK YOUR TRACE LISTS -------------------------
required_cols = {TIME_COL, SIG_COL}
for name, lst in [("ctcf_traces", 'ctcf_traces' in globals()),
                  ("pds5a_traces", 'pds5a_traces' in globals()),
                  ("coh_traces",   'coh_traces'   in globals()),
                  ("pc_traces",    'pc_traces'    in globals())]:
    if not lst:
        raise RuntimeError(f"{name} not defined. Define your trace DataFrame lists before running.")

for name, traces in [("CTCF", ctcf_traces), ("PDS5A", pds5a_traces), ("Cohesin", coh_traces), ("PDS5A+Coh", pc_traces)]:
    if len(traces)==0:
        print(f"Warning: no traces in {name}")
    else:
        missing = required_cols - set(traces[0].columns)
        if missing:
            raise ValueError(f"{name} traces missing columns: {missing}")

# ---------------------- SCORING (ALLOWED) ------------------------------
scores_ctcf  = score_traces_against_allowed(ctcf_traces,  model_dict, ["CTCF"],                              "ctcf_only",  outdir=OUTROOT)
scores_pds5a = score_traces_against_allowed(pds5a_traces, model_dict, ["CTCF","PDS5A"],                      "pds5a",      outdir=OUTROOT)
scores_coh   = score_traces_against_allowed(coh_traces,   model_dict, ["CTCF","Cohesin"],                     "cohesin",    outdir=OUTROOT)
scores_pc    = score_traces_against_allowed(pc_traces,    model_dict, ["CTCF","PDS5A","Cohesin","PDS5A+Coh"], "pds5a_coh",  outdir=OUTROOT)

for name, df in [("CTCF", scores_ctcf), ("PDS5A", scores_pds5a), ("Cohesin", scores_coh), ("PDS5A+Coh", scores_pc)]:
    print(f"{name} best-model counts:\n{df['best_model'].value_counts()}\n")

# -------------- REGIME ANNOTATION + PLOTS (ALLOWED + STABLE) ----------
summ_ctcf = run_condition_annotation_multiclass(
    "CTCF", ctcf_traces, model_dict, ["CTCF"],
    outdir=os.path.join(OUTROOT, "annot_ctcf"),
    win=WIN, step=STEP, delta_ll_switch=DELTA_LL_SWITCH, min_windows=MIN_WINDOWS_RUN
)

summ_pds5a = run_condition_annotation_multiclass(
    "PDS5A", pds5a_traces, model_dict, ["CTCF","PDS5A"],
    outdir=os.path.join(OUTROOT, "annot_pds5a"),
    win=WIN, step=STEP, delta_ll_switch=DELTA_LL_SWITCH, min_windows=MIN_WINDOWS_RUN
)

summ_coh = run_condition_annotation_multiclass(
    "Cohesin", coh_traces, model_dict, ["CTCF","Cohesin"],
    outdir=os.path.join(OUTROOT, "annot_cohesin"),
    win=WIN, step=STEP, delta_ll_switch=DELTA_LL_SWITCH, min_windows=MIN_WINDOWS_RUN
)

summ_pc = run_condition_annotation_multiclass(
    "PDS5A_Coh", pc_traces, model_dict, ["CTCF","PDS5A","Cohesin","PDS5A+Coh"],
    outdir=os.path.join(OUTROOT, "annot_pds5a_coh"),
    win=WIN, step=STEP, delta_ll_switch=DELTA_LL_SWITCH, min_windows=MIN_WINDOWS_RUN
)

all_summary = pd.concat([summ_ctcf, summ_pds5a, summ_coh, summ_pc], ignore_index=True)
all_summary.to_csv(os.path.join(OUTROOT, "all_conditions_regime_summary.csv"), index=False)
print(f"Saved → {os.path.join(OUTROOT, 'all_conditions_regime_summary.csv')}")

```

```python
import os
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from hmmlearn.hmm import GaussianHMM
from scipy.optimize import curve_fit

# ------------ columns (non-bp version) ------------
TIME_COL = "time (seconds)"
SIG_COL  = "force_corrected (pN)"

# ------------------ helpers ------------------
def to_col(x):
    x = np.asarray(x, dtype=float).reshape(-1,1)
    return x[~np.isnan(x[:,0])].reshape(-1,1)

def viterbi_decode(model: GaussianHMM, X):
    return model.predict(X)

def get_force_desc_order(model: GaussianHMM):
    """Indices of original states sorted by emission mean (desc)."""
    return list(np.argsort(model.means_.ravel())[::-1])

def transition_counts_raw(states, K):
    mat = np.zeros((K, K), dtype=float)
    if len(states) > 1:
        a = states[:-1]; b = states[1:]
        np.add.at(mat, (a, b), 1.0)
    return mat

def offdiag_transition_probs(counts):
    """Zero diagonal, renormalize each row over off-diagonals, return probs with NaN on diag."""
    K = counts.shape[0]
    off = counts.copy()
    np.fill_diagonal(off, 0.0)
    row_sums = off.sum(axis=1, keepdims=True)
    with np.errstate(divide='ignore', invalid='ignore'):
        probs = np.divide(off, row_sums, where=row_sums>0)
    probs[~np.isfinite(probs)] = 0.0
    np.fill_diagonal(probs, np.nan)
    return probs

def state_occupancies(states, K):
    counts = np.bincount(states, minlength=K).astype(float)
    total = counts.sum()
    return counts / total if total > 0 else np.full(K, np.nan)

def state_force_means(states, force):
    df = pd.DataFrame({'state': states, 'force': force})
    return df.groupby('state')['force'].mean().to_dict()

def compute_state_durations_seconds(states, times):
    """
    Given per-sample 'states' and matching 'times' (seconds), return dict: state -> list of dwell durations (s).
    """
    dur = {}
    if len(states) == 0 or len(times) == 0:
        return dur
    cur = states[0]
    start_t = times[0]
    for i in range(1, len(states)):
        if states[i] != cur:
            dt = times[i-1] - start_t
            if np.isfinite(dt) and dt > 0:
                dur.setdefault(cur, []).append(float(dt))
            cur = states[i]
            start_t = times[i]
    # close last dwell
    dt = times[-1] - start_t
    if np.isfinite(dt) and dt > 0:
        dur.setdefault(cur, []).append(float(dt))
    return dur

# ---- lifetime fitting (CDF) ----
def exponential_cdf(x, b):
    return 1 - np.exp(-b * x)

def fit_tau_from_durations(durations_s, max_s=5.0):
    d = np.array(durations_s, dtype=float)
    d = d[(d > 0) & np.isfinite(d) & (d < max_s)]
    if d.size < 3:
        return np.nan, np.nan, (None, None, None)

    num_bins = min(50, int(np.sqrt(d.size)))
    hist, bin_edges = np.histogram(d, bins=max(num_bins, 5), density=False)
    weights = 1.0 / np.sqrt(hist + 1e-10)

    bin_centers = 0.5*(bin_edges[:-1] + bin_edges[1:])
    cdf = np.cumsum(hist) / np.sum(hist)

    # prepend (0,0)
    bin_centers = np.insert(bin_centers, 0, 0.0)
    cdf = np.insert(cdf, 0, 0.0)
    weights = np.insert(weights, 0, weights[0])

    try:
        popt, pcov = curve_fit(exponential_cdf, bin_centers, cdf, p0=(1.0,),
                               sigma=weights, absolute_sigma=True, maxfev=20000)
        b = float(popt[0])
        if b <= 0 or not np.isfinite(b):
            return np.nan, np.nan, (None, None, None)
        tau_ms = (1.0 / b) * 1000.0
        tau_err_ms = (np.sqrt(np.diag(pcov))[0] / (b**2)) * 1000.0 if np.isfinite(pcov).all() else np.nan
        fitted = exponential_cdf(bin_centers, b)
        return tau_ms, tau_err_ms, (bin_centers*1000.0, cdf, fitted)
    except Exception:
        return np.nan, np.nan, (None, None, None)

# ------------- pure-segment analysis -------------
def analyse_pure_segments_with_lifetimes(summary_csv, model_path, pure_label,
                                         time_col=TIME_COL, force_col=SIG_COL):
    """
    Returns dict with keys:
      'order' (state indices desc by force),
      'occ', 'force', 'trans_off' (KxK with diag NaN),
      'durations' -> list-of-lists per canonical state,
      'taus_ms' and 'tau_err_ms' (per canonical state)
    """
    if not os.path.exists(summary_csv):
        print(f"[warn] Missing summary: {summary_csv}")
        return None
    if not os.path.exists(model_path):
        print(f"[warn] Missing model: {model_path}")
        return None

    summ = pd.read_csv(summary_csv)
    if "annotated_csv" not in summ.columns:
        print(f"[warn] No 'annotated_csv' in {summary_csv}")
        return None

    model = joblib.load(model_path)
    K = model.n_components
    order = get_force_desc_order(model)  # original indices sorted by mean force desc

    # accumulators
    occ_counts = np.zeros(K, dtype=float)
    total_len = 0
    force_vals = [[] for _ in range(K)]
    counts_sum = np.zeros((K, K), dtype=float)
    durations_by_state = {s: [] for s in range(K)}  # original indices

    for _, row in summ.iterrows():
        ann_path = row.get('annotated_csv')
        if not isinstance(ann_path, str) or not os.path.exists(ann_path):
            continue
        ann = pd.read_csv(ann_path)

        if force_col not in ann.columns or time_col not in ann.columns or "regime" not in ann.columns:
            continue

        mask = (ann['regime'] == pure_label) & ann[force_col].notna() & ann[time_col].notna()
        if not mask.any():
            continue

        force_seg = ann.loc[mask, force_col].to_numpy(dtype=float)
        time_seg  = ann.loc[mask, time_col].to_numpy(dtype=float)
        if force_seg.size < 2:
            continue

        X = to_col(force_seg)
        z = viterbi_decode(model, X)  # original indices

        # occupancies
        occ_counts += np.bincount(z, minlength=K)
        total_len += z.size

        # forces by state
        for s_idx in range(K):
            vals = force_seg[z == s_idx]
            if vals.size:
                force_vals[s_idx].append(vals)

        # transitions
        counts_sum += transition_counts_raw(z, K)

        # durations (seconds) per state
        dur_dict = compute_state_durations_seconds(z, time_seg)
        for s_idx, ds in dur_dict.items():
            durations_by_state[s_idx].extend(ds)

    if total_len == 0:
        print(f"[warn] No pure {pure_label} samples found in {summary_csv}")
        return None

    # aggregate
    occ = occ_counts / total_len
    mean_force = {s: (np.concatenate(force_vals[s]).mean() if len(force_vals[s]) else np.nan) for s in range(K)}
    trans_off = offdiag_transition_probs(counts_sum)

    # reorder to canonical order (1..K, where 1 = highest force)
    occ_ord   = occ[order]
    force_ord = np.array([mean_force.get(s, np.nan) for s in order])
    trans_ord = trans_off[np.ix_(order, order)]
    np.fill_diagonal(trans_ord, np.nan)
    durations_ord = [durations_by_state[s] for s in order]

    # fit taus per state (ms)
    taus_ms, tau_err_ms = [], []
    for dlist in durations_ord:
        tau, tau_err, _ = fit_tau_from_durations(dlist, max_s=5.0)
        taus_ms.append(tau); tau_err_ms.append(tau_err)

    return dict(
        K=K, order=order,
        occ=occ_ord,
        force=force_ord,
        trans_off=trans_ord,  # diag NaN, rows renormed over off-diagonals
        durations=durations_ord,
        taus_ms=np.array(taus_ms),
        tau_err_ms=np.array(tau_err_ms),
    )

# ----------------- pick model files (non-bp) -----------------
def choose_model(path_emissions_refit, path_transitions_refined):
    """Prefer emissions_refit if present, else fallback to transitions_refined."""
    if os.path.exists(path_emissions_refit):
        return path_emissions_refit
    return path_transitions_refined

conditions_info = [
    ("CTCF",
     "new_models/annot_ctcf/CTCF_summary.csv",
     choose_model("new_models/ctcf_emissions_refit.pkl",
                  "new_models/ctcf_transitions_refined.pkl"),
     "CTCF"),
    ("PDS5A",
     "new_models/annot_pds5a/PDS5A_summary.csv",
     choose_model("new_models/pds5a_emissions_refit.pkl",
                  "new_models/pds5a_transitions_refined.pkl"),
     "PDS5A"),
    ("Cohesin",
     "new_models/annot_cohesin/Cohesin_summary.csv",
     choose_model("new_models/cohesin_emissions_refit.pkl",
                  "new_models/cohesin_transitions_refined.pkl"),
     "Cohesin"),
    ("PDS5A+Coh",
     "new_models/annot_pds5a_coh/PDS5A_Coh_summary.csv",
     choose_model("new_models/pds5a_cohesin_emissions_refit.pkl",
                  "new_models/cohesin_pds5a_transitions_refined.pkl"),
     "PDS5A+Coh"),
]

# ----------------- run for your four conditions -----------------
results = {}
for name, summ_csv, model_path, pure_label in conditions_info:
    results[name] = analyse_pure_segments_with_lifetimes(summ_csv, model_path, pure_label,
                                                         time_col=TIME_COL, force_col=SIG_COL)

# ----------------- build data frames for grouped plots -----------------
results = {k:v for k,v in results.items() if v is not None}
conditions = list(results.keys())
K = 4
states_ticks = list(range(1, K+1))  # 1=highest force ... 4=lowest

def collect_by_state(key):
    # returns DataFrame rows=State(1..K), cols=Condition
    data = {}
    for cond in conditions:
        arr = results[cond][key]
        # ensure numpy array
        data[cond] = np.asarray(arr, dtype=float)
    df = pd.DataFrame(data, index=states_ticks)
    df.index.name = "State"
    return df

occ_df   = collect_by_state('occ')
force_df = collect_by_state('force')
tau_df   = collect_by_state('taus_ms')
tauerr_df= collect_by_state('tau_err_ms')

# ----------------- grouped bar plots (x=State, color=Condition) -----------------
def grouped_bar(df, ylabel, title, yerr_df=None):
    if df.empty:
        print(f"[warn] nothing to plot: {title}")
        return
    x = np.arange(len(df.index))
    ncond = len(df.columns)
    width = 0.8 / max(ncond, 1)
    fig, ax = plt.subplots(figsize=(8,3.4))
    for i, cond in enumerate(df.columns):
        vals = df[cond].values
        errs = (yerr_df[cond].values if (yerr_df is not None and cond in yerr_df.columns) else None)
        ax.bar(x + i*width - 0.4 + width/2, vals, width, label=cond, yerr=errs, capsize=3)
    ax.set_xticks(x); ax.set_xticklabels(df.index.astype(str))
    ax.set_xlabel("State (1 = highest force → 4 = lowest)")
    ax.set_ylabel(ylabel); ax.set_title(title)
    ax.legend(title="Condition", ncol=min(ncond, 4), fontsize=8)
    fig.tight_layout(); plt.show()

grouped_bar(occ_df,   "Occupancy fraction", "Occupancy by state (pure segments)")
grouped_bar(force_df, "Mean force (pN)",    "Mean force by state (pure segments)")
grouped_bar(tau_df,   "Lifetime τ (ms)",    "Exponential fit τ by state (pure segments)", yerr_df=tauerr_df)

# ----------------- transition heatmaps (off-diagonal only, %; diag blank) -----------------
for cond in conditions:
    trans = results[cond]['trans_off']  # already reordered, diag NaN
    fig, ax = plt.subplots(figsize=(4.2,3.6))
    mask = np.isnan(trans)
    sns.heatmap(trans*100.0, mask=mask, annot=True, fmt=".1f", cmap="Blues",
                cbar_kws={'label': '%'}, ax=ax, vmin=0, vmax=100)
    ax.set_title(f"Transition matrix (off-diagonal %) — {cond}")
    ax.set_xlabel("To state (1 = highest → 4 = lowest)")
    ax.set_ylabel("From state")
    ax.set_xticklabels(states_ticks); ax.set_yticklabels(states_ticks, rotation=0)
    fig.tight_layout(); plt.show()
```
